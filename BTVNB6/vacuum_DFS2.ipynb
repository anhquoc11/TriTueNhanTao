```python
import random
import copy 

a = [
    [1, 1, 1, 1],
    [1, 2, 1, 1],
    [1, 1, 2, 1],
    [1, 1, 1, 1]
]

GOAL = copy.deepcopy(a)
for t in range(4):
    for z in range(4):
        if GOAL[t][z] != 2: GOAL[t][z] = 0
        
i = random.randint(0, 3)
j = random.randint(0, 3)

while a[i][j] == 2:
    i = random.randint(0, 3)
    j = random.randint(0, 3)
```

```python

class Node():
    def __init__(self,i, j, state, action, parent, cost):
        self.i = i
        self.j = j
        self.STATE = state
        self.ACTION = action
        self.PARENT = parent
        self.COST = cost
```

```python

def get_actions(i,j):
    result = []
    if i > 0: result.append("UP")
    if i < 3: result.append("DOWN")
    if j > 0: result.append("LEFT")
    if j < 3: result.append("RIGHT")
    return result
```

```python
def remove_block(i, j, move, x):
    result = move.copy()
    if "UP" in result and x[i-1][j] == 2: result.remove("UP")
    if "DOWN" in result and x[i+1][j] == 2: result.remove("DOWN")
    if "LEFT" in result and x[i][j-1] == 2: result.remove("LEFT")
    if "RIGHT" in result and x[i][j+1] == 2: result.remove("RIGHT")
    return result
```

```python
def next_pos(i,j,x):
    if x == "UP": return i-1,j
    if x == "DOWN": return i+1,j
    if x == "RIGHT": return i,j+1
    if x == "LEFT": return i,j-1
```

```python
def print_board(board):
    for row in board:
        print(row)
    print()

print(f"Vị trí máy hút bụi ban đầu: {i,j}")
print("Sàn nhà ban đầu")
print_board(a)
```

```python

stack = []
reached = []
start_state = copy.deepcopy(a)
start_state[i][j] = 0
if start_state == GOAL:
    print("SÀN NHÀ ĐÃ ĐƯỢC DỌN SẠCH")
    print_board(GOAL)
    exit()
stack.append(Node(i,j,start_state,None,None,0))
reached.append((i, j, tuple(tuple(row) for row in start_state)))

find = False
while stack:
    cur = stack.pop()
    if cur.STATE == GOAL: 
        end = cur
        find = True
        break
    move_list = remove_block(cur.i,cur.j,get_actions(cur.i,cur.j),cur.STATE)
    for action in move_list:
        state_child = copy.deepcopy(cur.STATE)
        i_child , j_child = next_pos(cur.i,cur.j,action)
        state_child[i_child][j_child] = 0
        child = Node(i_child,j_child,state_child,action,cur,cur.COST + 1)
        state_tuple = tuple(tuple(row) for row in state_child)
        if (i_child,j_child,state_tuple) not in reached:
            stack.append(child)
            reached.append((i_child,j_child,state_tuple))
result = []
if find:
    result.append(end)
    print(f"SÀN NHÀ ĐƯỢC DỌN SẠCH SAU {end.COST} BƯỚC")
    while end.PARENT != None:
        end = end.PARENT
        result.insert(0,end)
    for temp in result:
        if temp.ACTION: print(temp.ACTION)
    print_board(GOAL)
else:
    print("SÀN NHÀ KHÔNG ĐƯỢC DỌN SẠCH")
    


```
