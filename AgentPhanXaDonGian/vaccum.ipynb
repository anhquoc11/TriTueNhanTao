PEAS của máy hút bụi 

Performance (P): làm sạch bụi hiệu quả, ít tốn điện, di chuyển nhanh, tránh vật cản.

Environment (E): phòng, sàn, bụi bẩn, tường, bàn ghế vật cản.

Actuators (A): bánh xe di chuyển, chổi/ quạt hút bụi, tay quay đổi hướng.

Sensors (S): cảm biến phát hiện bụi, camera, cảm biến phát hiện va chạm.

```python
import random

a = [
    [1, 0, 1, 1],
    [0, 1, 0, 1],
    [1, 1, 0, 0],
    [0, 1, 1, 0]
]

i = random.randint(0, 3)
j = random.randint(0, 3)

print(f"Vị trí máy hút bụi ban đầu: {i,j}")
print("Sàn nhà ban đầu")
for z in a:
    print(z)
```

```python
def move(i,j):
    result = []
    if i > 0: result.append("UP")
    if j > 0: result.append("LEFT")
    if i < 3: result.append("DOWN")
    if j < 3: result.append("RIGHT")
    return result
```

```python

def hehe(i,j,x):
    if x == "UP": return i-1,j
    if x == "DOWN": return i+1,j
    if x == "RIGHT": return i,j+1
    if x == "LEFT": return i,j-1
```

```python
while(True):
    old_a = [row[:] for row in a]
    b = [row[:] for row in a]
    for temp in range(5):
        if b[i][j] == 1: b[i][j] = 0
        huong = random.choice(move(i,j))
        print(huong)
        i,j = hehe(i,j,huong)
    if b == old_a:
        break
    a = [row[:] for row in b]
print("Sàn nhà sau khi hút bụi")
for z in a:
    print(z)
```
