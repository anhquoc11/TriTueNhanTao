```python
GOAL = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 0]
]
state = [
    [1, 2, 3],
    [4, 0, 6],
    [7, 5, 8]
]
```

```python
def find_zero(board):
    for i in range(3):
        for j in range(3):
            if board[i][j] == 0:
                return i, j
            
```

```python
def simple_reflex_rule(board):
    i, j = find_zero(board)
    if i > 0:
        return "UP"
    if j > 0:
        return "LEFT"
    if i < 2:
        return "DOWN"
    return "RIGHT"
```

```python
def move(board, action):
    i, j = find_zero(board)

    if action == "UP":
        board[i][j], board[i-1][j] = board[i-1][j], board[i][j]

    elif action == "DOWN":
        board[i][j], board[i+1][j] = board[i+1][j], board[i][j]

    elif action == "LEFT":
        board[i][j], board[i][j-1] = board[i][j-1], board[i][j]

    elif action == "RIGHT":
        board[i][j], board[i][j+1] = board[i][j+1], board[i][j]
```

```python
def print_board(board):
    for row in board:
        print(row)
    print()
```

```python
steps = 10

for step in range(steps):
    print("Step:", step)
    print_board(state)

    if state == GOAL:
        print("Đạt goal!")
        break
    action = simple_reflex_rule(state)
    move(state, action)
```
