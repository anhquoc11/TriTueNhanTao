```python
import random
import copy
GOAL = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 0]
]
state = [
    [1, 2, 3],
    [4, 0, 6],
    [7, 5, 8]
]
state_old = []
state_old.append(copy.deepcopy(state))
```

```python
def find_zero(board):
    for i in range(3):
        for j in range(3):
            if board[i][j] == 0:
                return i, j
```

```python
def based_reflex_agent(board):
    i, j = find_zero(board)
    result = []
    if i > 0: result.append("UP")
    if i < 2: result.append("DOWN")
    if j > 0: result.append("LEFT")
    if j < 2: result.append("RIGHT")
    return result
```

```python
def move(board, action):
    i, j = find_zero(board)
    newboard = copy.deepcopy(board)
    if action == "UP":
        newboard[i][j], newboard[i-1][j] = newboard[i-1][j], newboard[i][j]

    elif action == "DOWN":
        newboard[i][j], newboard[i+1][j] = newboard[i+1][j], newboard[i][j]

    elif action == "LEFT":
        newboard[i][j], newboard[i][j-1] = newboard[i][j-1], newboard[i][j]

    elif action == "RIGHT":
        newboard[i][j], newboard[i][j+1] = newboard[i][j+1], newboard[i][j]
    return newboard
```

```python
def model(StateOld,board):
    result = []
    for i in based_reflex_agent(board):
        new_state = move(board,i)
        if new_state not in StateOld: result.append(i)
    return result
```

```python
def print_board(board):
    for row in board:
        print(row)
    print()
```

```python
step = 0
while True:
    huong = model(state_old,state)
    if len(huong) == 0:
        print("Quay về trạng thái cũ")
        break
    dichuyen = random.choice(huong)
    state = move(state,dichuyen)
    state_old.append(copy.deepcopy(state))
    print(step)
    print_board(state)
    step+=1
    print(dichuyen)
    if state == GOAL:
        print("Đạt kết quả goal")
        break
```
