```python
import random
import copy

a = [
    [1, 0, 2, 1, 0],
    [2, 1, 0, 2, 1],
    [0, 2, 1, 0, 2],
    [1, 1, 2, 0, 1],
    [2, 0, 1, 2, 0]
]
visited = [[0 for _ in range(5)] for _ in range(5)]
i = random.randint(0, 4)
j = random.randint(0, 4)
while a[i][j] == 2:
    i = random.randint(0, 4)
    j = random.randint(0, 4)
print(f"Vị trí máy hút bụi ban đầu: {i,j}")
print("Sàn nhà ban đầu")
for z in a:
    print(z)
```

```python
def moves(i,j):
    result = []
    if i > 0: result.append("UP")
    if j > 0: result.append("LEFT")
    if i < 4: result.append("DOWN")
    if j < 4: result.append("RIGHT")
    return result
```

```python
def next_pos(i,j,x):
    if x == "UP": return i-1,j
    if x == "DOWN": return i+1,j
    if x == "RIGHT": return i,j+1
    if x == "LEFT": return i,j-1
```

```python
def remove_block(i,j,move, x):
    if "UP" in move and x[i-1][j] == 2:  move.remove("UP")
    if "DOWN" in move and x[i+1][j] == 2: move.remove("DOWN")
    if "LEFT" in move and x[i][j-1] == 2: move.remove("LEFT")
    if "RIGHT" in move and x[i][j+1] == 2: move.remove("RIGHT")
    return move
```

```python
def dirty(i,j,move,x):
    temp = []
    if "UP" in move and x[i-1][j] == 1:  temp.append("UP")
    if "DOWN" in move and x[i+1][j] == 1: temp.append("DOWN")
    if "LEFT" in move and x[i][j-1] == 1: temp.append("LEFT")
    if "RIGHT" in move and x[i][j+1] == 1: temp.append("RIGHT")
    return temp
```

```python
def Unvisited(i,j,move,x):
    temp = []
    if "UP" in move and visited[i-1][j] == 0: temp.append("UP")
    if "DOWN" in move and visited[i+1][j] == 0: temp.append("DOWN")
    if "LEFT" in move and visited[i][j-1] == 0: temp.append("LEFT")
    if "RIGHT" in move and visited[i][j+1] == 0: temp.append("RIGHT")
    return temp
```

```python
while(True):
    old_a = [row[:] for row in a]
    b = [row[:] for row in a]
    for temp in range(5):
        if b[i][j] == 1: b[i][j] = 0
        visited[i][j] = 1
        moves_list = remove_block(i,j,moves(i,j),b)
        if len(moves_list) == 0:
            print("Không thể di chuyển bởi vật cản")
            break
        dirty_moves = dirty(i,j,moves_list,b)
        if len(dirty_moves) != 0:
            huong = random.choice(dirty_moves)
        else:
            unvisited_moves = Unvisited(i,j,moves_list,b)
            if len(unvisited_moves) != 0:
                huong = random.choice(unvisited_moves)
            else: huong = random.choice(moves_list)
        print(huong)
        i,j = next_pos(i,j,huong)
    if b == old_a:
        break
    a = [row[:] for row in b]
print("Sàn nhà sau khi hút bụi")
for z in a:
    print(z)
```
