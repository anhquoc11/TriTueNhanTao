```python

import copy
GOAL = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 0]
]
state = [
    [1, 2, 3],
    [4, 0, 6],
    [7, 5, 8]
]
```

```python
class Node:
    def __init__(self, state, action, parent, cost):
        self.STATE = state
        self.ACTION = action
        self.PARENT = parent
        self.COST = cost
```

```python
def find_zero(board):
    for i in range(3):
        for j in range(3):
            if board[i][j] == 0:
                return i, j
            
```

```python
def get_actions(board):
    i, j = find_zero(board)
    result = []
    if i > 0: result.append("UP")
    if i < 2: result.append("DOWN")
    if j > 0: result.append("LEFT")
    if j < 2: result.append("RIGHT")
    return result
```

```python
def move(board, action):
    i, j = find_zero(board)
    newboard = copy.deepcopy(board)
    if action == "UP":
        newboard[i][j], newboard[i-1][j] = newboard[i-1][j], newboard[i][j]

    elif action == "DOWN":
        newboard[i][j], newboard[i+1][j] = newboard[i+1][j], newboard[i][j]

    elif action == "LEFT":
        newboard[i][j], newboard[i][j-1] = newboard[i][j-1], newboard[i][j]

    elif action == "RIGHT":
        newboard[i][j], newboard[i][j+1] = newboard[i][j+1], newboard[i][j]
    return newboard
```

```python
def print_board(board):
    for row in board:
        print(row)
    print()
```

```python
if state == GOAL:
    print("TRÒ CHƠI ĐÃ ĐƯỢC GIẢI")
    exit()
queue = []
reached = []
queue.append(Node(state,None,None,0))
reached.append(state)
find = False
while len(queue) != 0:
    if find: break
    temp = queue.pop(0)
    actions_list = get_actions(temp.STATE)
    for i in actions_list:
        new_state = move(temp.STATE,i)
        children = Node(new_state,i,temp,temp.COST + 1)
        if new_state == GOAL:
            end = children
            find = True
            break
        if new_state not in reached: 
            queue.append(children)
            reached.append(new_state)
if find:
    result = []
    result.append(end)
    print(f"TRÒ CHƠI HOÀN THÀNH SAU {end.COST} BƯỚC")
    while end.PARENT != None:
        end = end.PARENT
        result.insert(0,end)
    for i in result:
        print(i.ACTION)
        print_board(i.STATE)
    print("ĐÃ HOÀN THÀNH TRÒ CHƠI")
else:
    print("TRÒ CHƠI KHÔNG THỂ GIẢI")
    
```
